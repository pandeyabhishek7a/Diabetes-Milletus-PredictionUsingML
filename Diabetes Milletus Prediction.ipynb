# Taking Care of missing Data

```python
df.isnull().sum()
```

```python
x = df.iloc[:, 0:8].values
x
```

```python
y = df.iloc[:, 8:].values
y
```

# Splitting Data into Train and Test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

# Scaling of Data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

# Training and Testing the Model

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'entropy', random_state =0)
dt.fit(x_train, y_train)
```

```python
y_pred1 = dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred1, y_test)
```

# Using RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier 
rfc = RandomForestClassifier(criterion = 'gini', random_state = 0)
rfc.fit(x_train,y_train)
```

```python
y_pred = rfc.predict(x_test)
y_pred
```

```python
y_test
```

# Evaluation of Model

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```

# changing the Max Depth

```python
test_err = []

for i in range(1, 10):
    clf = RandomForestClassifier(max_depth = i, random_state = 10)
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    acc = accuracy_score(y_pred, y_test)
    error = 1-acc
    test_err.append(error)
```

```python
plt.plot(range(1,10), test_err)
plt.xlabel('Max_Depth')
plt.ylabel('Test_error')
plt.title('Finding value of Max Depth')
plt.show()
```

# Changing the max_leaf_nodes

```python
test_err = []

for i in range(2, 30):
    clf = RandomForestClassifier(max_leaf_nodes = i, random_state = 10)
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    acc = accuracy_score(y_pred, y_test)
    error = 1-acc
    test_err.append(error)
```

```python
plt.plot(range(2,30), test_err)
plt.xlabel('Max_Depth')
plt.ylabel('Test_error')
plt.title('Finding value of Max Depth')
plt.show()
```

```python
clf = RandomForestClassifier(criterion = 'gini', max_depth= 5, max_leaf_nodes = 9, random_state = 10)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
accuracy_score(y_pred, y_test)
```
